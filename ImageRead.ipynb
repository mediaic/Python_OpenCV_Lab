```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import cv2

# read image
img = cv2.imread('./test.jpg')

# check type
print("Data type: ", type(img))

# check shape
print("Image shape: ", np.array(img).shape)
print("dtype: ", np.array(img).dtype)

plt.imshow(img)
plt.show()
```

```python
from PIL import Image

# read image
img = Image.open('./test.jpg')

# check type
print("Data type: ", type(img))

# check shape
print("Image shape: ", np.array(img).shape)
print("dtype: ", np.array(img).dtype)

plt.imshow(img)
plt.show()
```
